1. **What is the difference between a function and a method in Python?**

  -> In Python, the main difference between a function and a method is how they are associated with objects:
    - Function: A function is an independent block of code that performs a specific task. It is defined using the def keyword and can be called without being tied to any object. Functions can accept arguments and return values.
    - Method: A method is a function that is defined inside a class and is associated with an object (instance of the class). It is called on an instance and can access or modify the instanceâ€™s attributes. Methods typically have self as their first parameter to refer to the instance (except for @staticmethod and @classmethod decorators)."

```python
#Example of a function:

# Example of a function:
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)
```

```python
# Example of a method:

# Example of a method using a class:
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

rect = Rectangle(4, 6)
print(rect.area())
```

6. Explain the concept of generators in Python and how they are defined.

  -> "In Python, a generator is a special type of function that produces values one at a time. Instead of generating and storing all the values at once (like a normal function), a generator yields a value and pauses its execution.\n",
    
    When the generator is called again, it picks up from where it left off and continues producing values one by one.,
    
  Key points about generators:
    
    -Memory-efficient: They don't store all values in memory, which is useful when dealing with large datasets.
    
    -On-demand: They produce values only when needed, one at a time.

    -Useful for large data: Great for looping through large datasets or streams without loading everything into memory.
  
  How to define a generator:
    A generator is defined like a regular function but uses the yield keyword to return a value instead of return.


```python
# example of generator :
def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count  # Yielding a value and pausing the function
        count += 1

# Using the generator
counter = count_up_to(3)

print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
print(next(counter))  # Output: 3
```

  7. What are the advantages of using generators over regular functions?
   
   -> Generators have several benefits, especially when dealing with large datasets or continuous data streams. Unlike regular functions, which return all results at once, generators yield values one by one.
    
    Advantages of using generators:
    
    Memory Efficiency:
    Generators only yield one value at a time and donâ€™t store the entire dataset in memory. This is helpful when working with large datasets.
    
    Faster Startup:

    Generators start working right away because they donâ€™t need to process the entire dataset at once. They yield values one by one as needed.

    
    Cleaner Code:
    Generators make your code simpler and easier to read. They are great when you only need to go through data without storing it all in memory.
    
    
   


```python
#Example of a generator for better clarity:

def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count  # Yielding one value at a time
        count += 1

counter = count_up_to(5)
print(next(counter))  # Output: 1
print(next(counter))  # Output: 2
```

  9. Explain the purpose and usage of the map() function in Python.

  -> In Python, the map() function is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new iterable with the results. Itâ€™s a quick way to perform the same operation on every element in a collection without writing a loop.

    
    Key points:
    
    Applies a function to each item in an iterable.
    
    Returns a new iterable (a map object) with the results of the function applied to each element.


```python
# example of map function:
# A simple function to add 2 to a number
def add_two(x):
    return x + 2

# List of numbers
numbers = [1, 2, 3, 4]

# Using map() to apply add_two function to each number
result = map(add_two, numbers)

# Convert map object to a list and print the result
print(list(result))  # Output: [3, 4, 5, 6]
```

  4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

  5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

# Example usage:
fib = FibonacciIterator(10)  # Get the first 10 Fibonacci numbers
for number in fib:
    print(number)
```

  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_2 (limit):
    for exponent in range(limit+1):
        yield exponent**2

for num in power_of_2(5):
    print(num)
```

  7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
# lets create a file so that python can read it.

with open ("data.txt", "w") as f:
    f.write("Hello PWSKILLS\n")
    f.write("Thank you for helping me\n")
    f.write("To create my actual carrier\n")

# lets create a function which will read this file :

def read_file_line_by_line (file_path):
    with open (file_path, "r") as my_file:
        for line in my_file:
            yield line.strip()

# Now lets test our generator function:

print ("Reading from the created file : ")

for line in read_file_line_by_line("data.txt"):
    print(line)
```
